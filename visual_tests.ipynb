```python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image

from autocrop import Cropper

loc = 'tests/data/macbeth.jpg'
c = Cropper(face_percent=56.7)
```

```python
c = Cropper(padding={'pad_top': 50,
                    'pad_bottom': 1000,
                    'pad_left': 50,
                    'pad_right': 50})

img_array = c.crop(loc)
plt.imshow(img_array)
plt.show()
```

```python
%%timeit
img_array = c.crop(loc)
```

```python
from glob import glob
faces = [f for f in glob('tests/data/*') if not f.endswith('md')]
```

```python
c = Cropper(width=650, height=500, face_percent=40)

for loc in faces:
    try:
        img_array = c.crop(loc)
        plt.imshow(img_array)
        plt.show()
    except:
        pass
```
